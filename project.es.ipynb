#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker, Icon, AwesomeIcon
import random

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (40.34, -3.7), zoom = 11, min_zoom = 1, max_zoom = 20, basemap=basemaps.OpenStreetMap.Mapnik)
```

```python
## Aquí: traza la coordenadas de los estados


## PON TU CÓDIGO AQUÍ:
grupo = ds_cinturonsur_copy.groupby('level5')[['latitude', 'longitude']].first()
diccionario = grupo.to_dict(orient='index')

colores = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'yellow', 'black']

for key, datos in diccionario.items():
    datos['latitude'] = float(datos['latitude'].replace(',', '.'))
    datos['longitude'] = float(datos['longitude'].replace(',', '.'))

for i, (key, datos) in enumerate(diccionario.items()):
    if datos["latitude"] and datos["longitude"]:
        color = colores[i % len(colores)]
        icono = AwesomeIcon(name= 'square', marker_color=color, icon_color='White')
        marcador = Marker(location=(datos["latitude"], datos["longitude"]), title=key, icon=icono)
        map.add_layer(marcador)
map
```
